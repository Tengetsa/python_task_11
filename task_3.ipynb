```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataekp_0.csv", delimiter = ",")
dfATA = df[df["section"] == "Авиамодельный спорт"]
dfAC = df[df["section"] == "Пауэрлифтинг"]
dfWA = df[df["section"] == "Плавание"]
```

```python
fig, ax = plt.subplots()
plt.stem(pd.to_datetime(dfATA.start), dfATA["stop"], "-*k")
plt.stem(pd.to_datetime(dfAC.start), dfAC["stop"], "-or")
plt.stem(pd.to_datetime(dfWA.start), dfWA["stop"], "-b")
plt.show()
```
